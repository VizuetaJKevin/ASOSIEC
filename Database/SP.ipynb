# [dbo].[sp_Actualizar_Perfil]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Actualizar_Perfil]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Crear el SP corregido
CREATE PROCEDURE [dbo].[sp_Actualizar_Perfil]
    @id INT,
    @nombre NVARCHAR(100),
    @apellido NVARCHAR(100),
    @email NVARCHAR(100),
    @telefono NVARCHAR(20) = NULL,
    @direccion NVARCHAR(500) = NULL,
    @fotoPerfil NVARCHAR(500) = NULL
AS
BEGIN
    -- CRÍTICO: NO usar SET NOCOUNT ON para que @@ROWCOUNT funcione
    -- SET NOCOUNT ON suprime el mensaje de filas afectadas
    
    BEGIN TRY
        -- Realizar el UPDATE
        UPDATE Usuario
        SET 
            nombre = @nombre,
            apellido = @apellido,
            email = @email,
            telefono = @telefono,
            direccion = @direccion,
            fotoPerfil = @fotoPerfil
        WHERE id = @id;
        
        -- El @@ROWCOUNT ahora será capturado por ExecuteNonQuery()
        -- No necesitamos RETURN porque ExecuteNonQuery() lee directamente las filas afectadas
        
    END TRY
    BEGIN CATCH
        -- En caso de error, lanzar la excepción para que C# la capture
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE()
        DECLARE @ErrorSeverity INT = ERROR_SEVERITY()
        DECLARE @ErrorState INT = ERROR_STATE()
        
        RAISERROR(@ErrorMessage, @ErrorSeverity, @ErrorState)
    END CATCH
END
GO
```

# [dbo].[sp_Aprobar_Devolucion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Aprobar_Devolucion]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- ============================================
-- STORED PROCEDURES PARA DEVOLUCIONES
-- Sistema de Gestión de Devoluciones con Emails
-- ============================================

-- ============================================
-- 1. SP: Aprobar Devolución
-- Este SP aprueba una devolución, actualiza su estado y registra auditoría
-- ============================================
CREATE   PROCEDURE [dbo].[sp_Aprobar_Devolucion]
    @devolucionId INT,
    @numero_seguimiento VARCHAR(100),
    @respondido_por INT,
    @respuesta_admin VARCHAR(500) = NULL,
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @estadoAprobadoId INT;
    DECLARE @ordenId INT;
    DECLARE @usuarioId INT;
    DECLARE @tipo_devolucion VARCHAR(20);
    
    BEGIN TRY
        BEGIN TRANSACTION;
        
        -- Obtener ID del estado "APROBADO"
        SELECT @estadoAprobadoId = id 
        FROM estado_devolucion 
        WHERE codigo = 'APROBADO';
        
        IF @estadoAprobadoId IS NULL
        BEGIN
            RAISERROR('Estado APROBADO no encontrado', 16, 1);
            ROLLBACK TRANSACTION;
            SET @accion = 0;
            RETURN;
        END
        
        -- Obtener datos de la devolución
        SELECT 
            @ordenId = ordenId,
            @usuarioId = usuarioId,
            @tipo_devolucion = tipo_devolucion
        FROM devolucion
        WHERE id = @devolucionId;
        
        -- Actualizar la devolución
        UPDATE devolucion
        SET 
            estadoDevolucionId = @estadoAprobadoId,
            numero_seguimiento = @numero_seguimiento,
            fecha_respuesta = GETDATE(),
            respondido_por = @respondido_por,
            respuesta_admin = @respuesta_admin
        WHERE id = @devolucionId;
        
        -- Registrar en auditoría
        INSERT INTO auditoria (
            entidad,
            accion,
            descripcion,
            usuarioId,
            fecha
        )
        VALUES (
            'devolucion',
            'APROBAR',
            'Devolución #' + CAST(@devolucionId AS VARCHAR) + ' aprobada. N° Seguimiento: ' + @numero_seguimiento,
            @respondido_por,
            GETDATE()
        );
        
        COMMIT TRANSACTION;
        SET @accion = 1;
        
        PRINT '✅ Devolución aprobada exitosamente';
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
            
        SET @accion = 0;
        PRINT '❌ Error al aprobar devolución: ' + ERROR_MESSAGE();
    END CATCH
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_Aprobar_Devolucion_SinEmails]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Aprobar_Devolucion_SinEmails]
    @devolucionId INT,
    @numero_seguimiento VARCHAR(100),
    @respondido_por INT,
    @respuesta_admin VARCHAR(500) = NULL,
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    BEGIN TRY
        BEGIN TRANSACTION;

        DECLARE @estado_aprobado_id INT;
        DECLARE @ordenId INT;
        DECLARE @total_devolucion DECIMAL(10,2);

        SELECT @estado_aprobado_id = id 
        FROM estado_devolucion 
        WHERE codigo = 'APROBADO';

        IF @estado_aprobado_id IS NULL
        BEGIN
            SET @accion = 0;
            ROLLBACK TRANSACTION;
            RAISERROR('Estado APROBADO no encontrado', 16, 1);
            RETURN;
        END

        SELECT @ordenId = ordenId
        FROM devolucion
        WHERE id = @devolucionId;

        IF @ordenId IS NULL
        BEGIN
            SET @accion = 0;
            ROLLBACK TRANSACTION;
            RAISERROR('Devolucion no encontrada', 16, 1);
            RETURN;
        END

        SELECT @total_devolucion = ISNULL(SUM(cantidad * precio_unitario), 0)
        FROM item_devolucion
        WHERE devolucionId = @devolucionId;

        UPDATE devolucion
        SET 
            estadoDevolucionId = @estado_aprobado_id,
            numero_seguimiento = @numero_seguimiento,
            fecha_respuesta = GETDATE(),
            respondido_por = @respondido_por,
            respuesta_admin = ISNULL(@respuesta_admin, 'Devolucion aprobada.')
        WHERE id = @devolucionId;

        UPDATE Orden
        SET total = total - @total_devolucion
        WHERE id = @ordenId;

        COMMIT TRANSACTION;
        SET @accion = 1;

    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
        SET @accion = 0;
        DECLARE @err NVARCHAR(4000) = ERROR_MESSAGE();
        PRINT 'Error: ' + @err;
        RAISERROR(@err, 16, 1);
    END CATCH
END
GO
```

# [dbo].[sp_Aprobar_Producto]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Aprobar_Producto]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Aprobar_Producto]
    @productoId INT,
    @adminId INT,
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        BEGIN TRANSACTION
        
        -- Verificar que el producto existe y está pendiente
        IF NOT EXISTS (SELECT 1 FROM producto WHERE id = @productoId AND estadoProductoId = 4)
        BEGIN
            SET @accion = 0;
            ROLLBACK TRANSACTION;
            PRINT '❌ El producto no existe o no está pendiente de aprobación';
            RETURN;
        END
        
        -- Actualizar producto a DISPONIBLE y marcar como aprobado
        UPDATE producto
        SET 
            estadoProductoId = 1,  -- DISPONIBLE
            aprobado = 1,
            fecha_aprobacion = GETDATE(),
            aprobado_por = @adminId,
            motivo_rechazo = NULL  -- Limpiar motivo de rechazo si existía
        WHERE id = @productoId;
        
        COMMIT TRANSACTION;
        SET @accion = 1;
        
        PRINT '✅ Producto ' + CAST(@productoId AS VARCHAR) + ' aprobado correctamente';
        
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
        
        SET @accion = 0;
        
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE();
        PRINT '❌ Error al aprobar producto: ' + @ErrorMessage;
        RAISERROR (@ErrorMessage, 16, 1);
    END CATCH
END
GO
```

# [dbo].[sp_Aprobar_Usuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Aprobar_Usuario]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_Aprobar_Usuario]
@usuarioId INT,
@adminId INT,
@accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        BEGIN TRANSACTION
        
        DECLARE @nombre NVARCHAR(100)
        DECLARE @apellido NVARCHAR(100)
        DECLARE @rolId INT
        
        SELECT 
            @nombre = nombre,
            @apellido = apellido,
            @rolId = rolId
        FROM usuario
        WHERE id = @usuarioId
        
        UPDATE usuario
        SET 
            aprobado = 1,
            fecha_aprobacion = GETDATE(),
            aprobado_por = @adminId,
            estadoUsuarioId = 1
        WHERE id = @usuarioId
        
        IF @rolId = 3
        BEGIN
            IF NOT EXISTS (SELECT 1 FROM vendedor WHERE usuarioId = @usuarioId)
            BEGIN
                INSERT INTO vendedor (usuarioId, estadoUsuarioId, nombre_comercial, descripcion, telefono, fecha_registro)
                VALUES (
                    @usuarioId,
                    1,
                    @nombre + ' ' + @apellido,
                    'Vendedor verificado',
                    '',
                    GETDATE()
                )
            END
            ELSE
            BEGIN
                UPDATE vendedor 
                SET estadoUsuarioId = 1,
                    nombre_comercial = @nombre + ' ' + @apellido
                WHERE usuarioId = @usuarioId
            END
        END
        
        COMMIT TRANSACTION
        SET @accion = 1
        
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION
        
        SET @accion = 0
        
        -- ✅ CORRECCIÓN: Capturar el mensaje antes de usarlo
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE()
        RAISERROR (@ErrorMessage, 16, 1)
    END CATCH
END
GO
```

# [dbo].[sp_comprobantes_pendientes]

```sql
/****** Object:  StoredProcedure [dbo].[sp_comprobantes_pendientes]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- ============================================
-- STORED PROCEDURE: sp_comprobantes_pendientes
-- Versión FINAL basada en estructura real de tablas
-- ============================================

CREATE PROCEDURE [dbo].[sp_comprobantes_pendientes]
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT 
        -- ✅ Identificadores (con alias para coincidir con TypeScript)
        c.id AS comprobante_id,
        c.ordenId AS orden_id,
        
        -- ✅ Datos del comprobante
        c.url_comprobante,
        c.numero_referencia,
        c.verificado,
        c.fecha_subida,
        
        -- ✅ Información de la orden
        o.token_orden,
        o.total AS monto_total,
        o.fecha AS fecha_orden,
        o.estadoOrdenId,
        
        -- ✅ Cliente (nombre concatenado)
        CONCAT(o.nombre, ' ', o.apellido) AS cliente_nombre,
        o.email AS cliente_email,
        
        -- ✅ Días pendientes calculados
        DATEDIFF(DAY, c.fecha_subida, GETDATE()) AS dias_pendiente
        
    FROM comprobante_pago c
    INNER JOIN orden o ON c.ordenId = o.id
    WHERE 
        c.verificado = 0              -- Solo no verificados
        AND o.estadoOrdenId = 3       -- Solo "COMPROBANTE_ENVIADO"
    ORDER BY c.fecha_subida DESC;
END
GO
```

# [dbo].[sp_Consultar_Auditoria]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Consultar_Auditoria]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Consultar_Auditoria]
    @fecha_inicio DATETIME2 = NULL,
    @fecha_fin DATETIME2 = NULL,
    @usuario_id INT = NULL,
    @entidad VARCHAR(100) = NULL,
    @entidad_id INT = NULL,
    @tipo_operacion VARCHAR(50) = NULL,
    @solo_exitosos BIT = NULL,
    @pagina INT = 1,
    @registros_por_pagina INT = 50
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Calcular offset
    DECLARE @offset INT = (@pagina - 1) * @registros_por_pagina;
    
    -- Consulta principal
    SELECT 
        id, fecha_evento, usuario_id, usuario_email, usuario_nombre, usuario_rol,
        tipo_operacion, entidad, entidad_id, accion, tabla_afectada,
        valores_anteriores, valores_nuevos,
        ip_address, user_agent, endpoint, metodo_http,
        exito, codigo_error, mensaje_error, datos_adicionales
    FROM AuditLog
    WHERE 
        (@fecha_inicio IS NULL OR fecha_evento >= @fecha_inicio)
        AND (@fecha_fin IS NULL OR fecha_evento <= @fecha_fin)
        AND (@usuario_id IS NULL OR usuario_id = @usuario_id)
        AND (@entidad IS NULL OR entidad = @entidad)
        AND (@entidad_id IS NULL OR entidad_id = @entidad_id)
        AND (@tipo_operacion IS NULL OR tipo_operacion = @tipo_operacion)
        AND (@solo_exitosos IS NULL OR exito = @solo_exitosos)
    ORDER BY fecha_evento DESC
    OFFSET @offset ROWS
    FETCH NEXT @registros_por_pagina ROWS ONLY;
    
    -- Retornar total de registros
    SELECT COUNT(*) as total_registros
    FROM AuditLog
    WHERE 
        (@fecha_inicio IS NULL OR fecha_evento >= @fecha_inicio)
        AND (@fecha_fin IS NULL OR fecha_evento <= @fecha_fin)
        AND (@usuario_id IS NULL OR usuario_id = @usuario_id)
        AND (@entidad IS NULL OR entidad = @entidad)
        AND (@entidad_id IS NULL OR entidad_id = @entidad_id)
        AND (@tipo_operacion IS NULL OR tipo_operacion = @tipo_operacion)
        AND (@solo_exitosos IS NULL OR exito = @solo_exitosos);
END
GO
```

# [dbo].[sp_consultar_productos]

```sql
/****** Object:  StoredProcedure [dbo].[sp_consultar_productos]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_consultar_productos]
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT 
        p.*,
        v.nombre_comercial AS vendedor_nombre,
        c.nombre_categoria,
        m.nombre_marca,
        ep.descripcion AS estado_nombre,
        admin.nombre + ' ' + admin.apellido AS aprobado_por_nombre
    FROM producto p
    LEFT JOIN vendedor v ON p.vendedorId = v.id
    LEFT JOIN categoria_producto c ON p.categoria_producto_Id = c.id
    LEFT JOIN marca_producto m ON p.marcaId = m.id
    INNER JOIN estado_producto ep ON p.estadoProductoId = ep.id
    LEFT JOIN usuario admin ON p.aprobado_por = admin.id
    ORDER BY p.id DESC;
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_Crear_Codigo_Recuperacion]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[sp_Crear_Codigo_Recuperacion]
    @email       VARCHAR(255),
    @codigo      VARCHAR(6),       -- Código numérico de 6 dígitos
    @ip_solicitud VARCHAR(45) = NULL
AS
BEGIN
    SET NOCOUNT ON;

    -- Invalidar cualquier código previo activo para ese email
    UPDATE Tokens_Recuperacion
    SET usado = 1, fecha_uso = GETDATE()
    WHERE email = @email AND usado = 0;

    -- Insertar nuevo código con expiración de 15 minutos
    INSERT INTO Tokens_Recuperacion (email, token, fecha_expiracion, ip_solicitud)
    VALUES (@email, @codigo, DATEADD(MINUTE, 15, GETDATE()), @ip_solicitud);

    SELECT 1 AS resultado;
END;
GO
```

# [dbo].[sp_dashboard_kpis]

```sql
/****** Object:  StoredProcedure [dbo].[sp_dashboard_kpis]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_dashboard_kpis]
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @VentasMesActual DECIMAL(18,2);
    DECLARE @VentasMesAnterior DECIMAL(18,2);
    DECLARE @CrecimientoPorc DECIMAL(10,2);
    
    -- Ventas del mes actual
    SELECT @VentasMesActual = ISNULL(SUM(total), 0)
    FROM Orden
    WHERE MONTH(fecha) = MONTH(GETDATE())
      AND YEAR(fecha) = YEAR(GETDATE())
      AND estadoOrdenId IN (4, 5);
    
    -- Ventas del mes anterior
    SELECT @VentasMesAnterior = ISNULL(SUM(total), 0)
    FROM Orden
    WHERE MONTH(fecha) = MONTH(DATEADD(MONTH, -1, GETDATE()))
      AND YEAR(fecha) = YEAR(DATEADD(MONTH, -1, GETDATE()))
      AND estadoOrdenId IN (4, 5);
    
    -- Calcular crecimiento
    IF @VentasMesAnterior > 0
        SET @CrecimientoPorc = ((@VentasMesActual - @VentasMesAnterior) / @VentasMesAnterior) * 100;
    ELSE
        SET @CrecimientoPorc = 0;
    
    SELECT 
        -- Ventas
        @VentasMesActual as ventas_mes_actual,
        @VentasMesAnterior as ventas_mes_anterior,
        @CrecimientoPorc as crecimiento_porcentaje,
        
        -- Órdenes
        (SELECT COUNT(*) FROM Orden WHERE estadoOrdenId IN (3, 4)) as ordenes_pendientes,
        (SELECT COUNT(*) FROM Orden WHERE estadoOrdenId = 5 
            AND MONTH(fecha) = MONTH(GETDATE()) 
            AND YEAR(fecha) = YEAR(GETDATE())) as ordenes_completadas_mes,
        (SELECT COUNT(*) FROM Orden) as total_ordenes,
        
        -- Productos
        (SELECT COUNT(*) FROM producto WHERE estadoProductoId IN (1, 2, 3)) as total_productos,
        (SELECT COUNT(*) FROM producto WHERE estadoProductoId = 2) as productos_sin_stock,
        
        -- Usuarios
        (SELECT COUNT(*) FROM usuario WHERE rolId = 2 AND estadoUsuarioId = 1) as total_clientes,
        (SELECT COUNT(*) FROM vendedor WHERE estadoUsuarioId = 1) as vendedores_activos,
        (SELECT COUNT(*) FROM usuario WHERE rolId = 3 AND estadoUsuarioId = 4) as solicitudes_pendientes
END
GO
```

# [dbo].[sp_dashboard_kpis_vendedor]

```sql
/****** Object:  StoredProcedure [dbo].[sp_dashboard_kpis_vendedor]    Script Date: 10/03/2026 7:13:33 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_dashboard_kpis_vendedor]
@vendedorId INT
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @VentasMesActual DECIMAL(18,2);
    DECLARE @VentasMesAnterior DECIMAL(18,2);
    DECLARE @CrecimientoPorc DECIMAL(10,2);
    
    -- Ventas del mes actual del vendedor
    SELECT @VentasMesActual = ISNULL(SUM(i.cantidad * p.precio_ahora), 0)
    FROM Items i
    INNER JOIN producto p ON p.id = i.productoId
    INNER JOIN Orden o ON o.id = i.ordenId
    WHERE p.vendedorId = @vendedorId
      AND MONTH(o.fecha) = MONTH(GETDATE())
      AND YEAR(o.fecha) = YEAR(GETDATE())
      AND o.estadoOrdenId IN (4, 5)
      AND i.estadoItemId = 2;
    
    -- Ventas del mes anterior del vendedor
    SELECT @VentasMesAnterior = ISNULL(SUM(i.cantidad * p.precio_ahora), 0)
    FROM Items i
    INNER JOIN producto p ON p.id = i.productoId
    INNER JOIN Orden o ON o.id = i.ordenId
    WHERE p.vendedorId = @vendedorId
      AND MONTH(o.fecha) = MONTH(DATEADD(MONTH, -1, GETDATE()))
      AND YEAR(o.fecha) = YEAR(DATEADD(MONTH, -1, GETDATE()))
      AND o.estadoOrdenId IN (4, 5)
      AND i.estadoItemId = 2;
    
    -- Calcular crecimiento
    IF @VentasMesAnterior > 0
        SET @CrecimientoPorc = ((@VentasMesActual - @VentasMesAnterior) / @VentasMesAnterior) * 100;
    ELSE
        SET @CrecimientoPorc = 0;
    
    SELECT 
        -- Ventas
        @VentasMesActual as ventas_mes_actual,
        @VentasMesAnterior as ventas_mes_anterior,
        @CrecimientoPorc as crecimiento_porcentaje,
        
        -- Productos
        (SELECT COUNT(*) FROM producto WHERE vendedorId = @vendedorId AND estadoProductoId IN (1, 2)) as total_productos,
        (SELECT COUNT(*) FROM producto WHERE vendedorId = @vendedorId AND estadoProductoId = 2) as productos_sin_stock,
        
        -- Órdenes (calculadas desde items)
        (SELECT COUNT(DISTINCT o.id) FROM Orden o
         INNER JOIN Items i ON i.ordenId = o.id
         INNER JOIN producto p ON p.id = i.productoId
         WHERE p.vendedorId = @vendedorId AND o.estadoOrdenId IN (3, 4)) as ordenes_pendientes,
        
        (SELECT COUNT(DISTINCT o.id) FROM Orden o
         INNER JOIN Items i ON i.ordenId = o.id
         INNER JOIN producto p ON p.id = i.productoId
         WHERE p.vendedorId = @vendedorId 
           AND o.estadoOrdenId = 5
           AND MONTH(o.fecha) = MONTH(GETDATE())
           AND YEAR(o.fecha) = YEAR(GETDATE())) as ordenes_completadas_mes,
        
        (SELECT COUNT(DISTINCT o.id) FROM Orden o
         INNER JOIN Items i ON i.ordenId = o.id
         INNER JOIN producto p ON p.id = i.productoId
         WHERE p.vendedorId = @vendedorId) as total_ordenes,
        
        -- Vendedor (para compatibilidad con la interfaz del admin)
        0 as total_clientes,
        1 as vendedores_activos,
        0 as solicitudes_pendientes
END
GO
```

# [dbo].[sp_Estadisticas_Auditoria]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Estadisticas_Auditoria]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Estadisticas_Auditoria]
    @fecha_inicio DATETIME2 = NULL,
    @fecha_fin DATETIME2 = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Si no se especifica rango, últimos 30 días
    IF @fecha_inicio IS NULL SET @fecha_inicio = DATEADD(DAY, -30, GETDATE());
    IF @fecha_fin IS NULL SET @fecha_fin = GETDATE();
    
    -- 1. Operaciones por tipo
    SELECT 
        tipo_operacion,
        COUNT(*) as cantidad,
        SUM(CASE WHEN exito = 1 THEN 1 ELSE 0 END) as exitosos,
        SUM(CASE WHEN exito = 0 THEN 1 ELSE 0 END) as fallidos
    FROM AuditLog
    WHERE fecha_evento BETWEEN @fecha_inicio AND @fecha_fin
    GROUP BY tipo_operacion;
    
    -- 2. Actividad por usuario
    SELECT TOP 10
        usuario_id,
        usuario_email,
        usuario_rol,
        COUNT(*) as total_operaciones,
        MAX(fecha_evento) as ultima_actividad
    FROM AuditLog
    WHERE fecha_evento BETWEEN @fecha_inicio AND @fecha_fin
        AND usuario_id IS NOT NULL
    GROUP BY usuario_id, usuario_email, usuario_rol
    ORDER BY total_operaciones DESC;
    
    -- 3. Entidades más modificadas
    SELECT TOP 10
        entidad,
        COUNT(*) as total_cambios,
        SUM(CASE WHEN tipo_operacion = 'INSERT' THEN 1 ELSE 0 END) as inserciones,
        SUM(CASE WHEN tipo_operacion = 'UPDATE' THEN 1 ELSE 0 END) as actualizaciones,
        SUM(CASE WHEN tipo_operacion = 'DELETE' THEN 1 ELSE 0 END) as eliminaciones
    FROM AuditLog
    WHERE fecha_evento BETWEEN @fecha_inicio AND @fecha_fin
    GROUP BY entidad
    ORDER BY total_cambios DESC;
    
    -- 4. Actividad por día
    SELECT 
        CAST(fecha_evento AS DATE) as fecha,
        COUNT(*) as operaciones,
        COUNT(DISTINCT usuario_id) as usuarios_activos
    FROM AuditLog
    WHERE fecha_evento BETWEEN @fecha_inicio AND @fecha_fin
    GROUP BY CAST(fecha_evento AS DATE)
    ORDER BY fecha DESC;
END
GO
```

# [dbo].[sp_Guardar_Configuracion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Guardar_Configuracion]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Guardar_Configuracion]
    @clave VARCHAR(100),
    @valor VARCHAR(500),
    @descripcion VARCHAR(255) = NULL,
    @tipo VARCHAR(20) = 'STRING',
    @activo BIT = 1,
    @modificado_por INT = NULL,
    @accion BIT OUTPUT  -- ✅ PARÁMETRO DE SALIDA
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        BEGIN TRANSACTION;
        
        -- Verificar si la configuración ya existe
        IF EXISTS (SELECT 1 FROM Configuracion WHERE clave = @clave)
        BEGIN
            -- ACTUALIZAR configuración existente
            UPDATE Configuracion
            SET 
                valor = @valor,
                descripcion = ISNULL(@descripcion, descripcion),
                tipo = @tipo,
                activo = @activo,  -- ✅ SIEMPRE actualizar activo
                fecha_modificacion = GETDATE(),
                modificado_por = @modificado_por
            WHERE clave = @clave;
            
            -- ✅ Marcar como exitoso
            SET @accion = 1;
            
            PRINT '✅ Configuración ACTUALIZADA: ' + @clave;
        END
        ELSE
        BEGIN
            -- INSERTAR nueva configuración
            INSERT INTO Configuracion (clave, valor, descripcion, tipo, activo, fecha_creacion, modificado_por)
            VALUES (@clave, @valor, @descripcion, @tipo, @activo, GETDATE(), @modificado_por);
            
            -- ✅ Marcar como exitoso
            SET @accion = 1;
            
            PRINT '✅ Configuración CREADA: ' + @clave;
        END
        
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        
        -- ❌ Marcar como fallido
        SET @accion = 0;
        
        PRINT '❌ Error al guardar configuración: ' + ERROR_MESSAGE();
        
        -- Re-lanzar el error
        THROW;
    END CATCH
END;
GO
```

# [dbo].[sp_Obtener_Estadisticas_Usuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Obtener_Estadisticas_Usuario]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Crear el procedimiento corregido
CREATE PROCEDURE [dbo].[sp_Obtener_Estadisticas_Usuario]
    @usuarioId INT
AS
BEGIN
    SET NOCOUNT ON;
    
    -- ============================================
    -- RESULTADO 1: Productos comprados en los últimos 7 días y total gastado
    -- ============================================
    SELECT 
        COUNT(DISTINCT i.productoId) as productosComprados7Dias,
        ISNULL(SUM(i.cantidad * p.precio_ahora), 0) as totalGastado
    FROM Orden o
    INNER JOIN Items i ON o.id = i.ordenId  -- ✅ CORREGIDO: "Item" → "Items"
    INNER JOIN producto p ON i.productoId = p.id
    WHERE o.usuarioId = @usuarioId
    AND o.fecha >= DATEADD(DAY, -7, GETDATE())
    AND o.estadoOrdenId IN (3, 4, 5)  -- Solo órdenes con comprobante o verificadas
    AND i.estadoItemId IN (1, 2);  -- En carrito o vendido

    -- ============================================
    -- RESULTADO 2: Compras por mes (últimos 6 meses)
    -- ============================================
    SELECT 
        FORMAT(o.fecha, 'MMM') as mes,
        COUNT(DISTINCT i.productoId) as cantidad,
        ISNULL(SUM(i.cantidad * p.precio_ahora), 0) as monto
    FROM Orden o
    INNER JOIN Items i ON o.id = i.ordenId  -- ✅ CORREGIDO: "Item" → "Items"
    INNER JOIN producto p ON i.productoId = p.id
    WHERE o.usuarioId = @usuarioId
    AND o.fecha >= DATEADD(MONTH, -6, GETDATE())
    AND o.estadoOrdenId IN (3, 4, 5)
    AND i.estadoItemId IN (1, 2)
    GROUP BY FORMAT(o.fecha, 'MMM'), MONTH(o.fecha)
    ORDER BY MONTH(o.fecha);
END
GO
```

# [dbo].[sp_productos_mas_vendidos]

```sql
/****** Object:  StoredProcedure [dbo].[sp_productos_mas_vendidos]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_productos_mas_vendidos]
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT TOP 10
        p.id as producto_id,
        p.nombre_producto,
        ISNULL(SUM(i.cantidad), 0) as cantidad_vendida,
        ISNULL(SUM(i.cantidad * p.precio_ahora), 0) as total_generado,
        p.url_Img as url_img
    FROM producto p
    LEFT JOIN Items i ON i.productoId = p.id 
        AND i.ordenId IS NOT NULL
        AND i.estadoItemId = 2 -- Solo items vendidos
    LEFT JOIN Orden o ON o.id = i.ordenId 
        AND o.estadoOrdenId IN (4, 5) -- Solo órdenes verificadas o completadas
    WHERE p.estadoProductoId IN (1, 2) -- Disponible o Agotado (no descontinuados)
    GROUP BY p.id, p.nombre_producto, p.url_Img
    HAVING SUM(i.cantidad) > 0
    ORDER BY cantidad_vendida DESC, total_generado DESC
END
GO
```

# [dbo].[sp_productos_vendedor_estadisticas]

```sql
/****** Object:  StoredProcedure [dbo].[sp_productos_vendedor_estadisticas]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_productos_vendedor_estadisticas]
@vendedorId INT
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT 
        p.id as producto_id,
        p.nombre_producto,
        p.stock as stock_actual,
        ISNULL(SUM(i.cantidad), 0) as cantidad_vendida,
        p.precio_ahora as precio_actual,
        ISNULL(SUM(i.cantidad * p.precio_ahora), 0) as total_generado
    FROM producto p
    LEFT JOIN Items i ON i.productoId = p.id AND i.estadoItemId = 2
    LEFT JOIN Orden o ON o.id = i.ordenId AND o.estadoOrdenId IN (4, 5)
    WHERE p.vendedorId = @vendedorId
      AND p.estadoProductoId IN (1, 2)
    GROUP BY p.id, p.nombre_producto, p.stock, p.precio_ahora
    ORDER BY cantidad_vendida DESC, total_generado DESC
END
GO
```

# [dbo].[sp_Rechazar_Devolucion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Rechazar_Devolucion]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- 2. SP: Rechazar Devolución
-- Este SP rechaza una devolución, actualiza su estado y registra auditoría
-- ============================================
CREATE   PROCEDURE [dbo].[sp_Rechazar_Devolucion]
    @devolucionId INT,
    @respondido_por INT,
    @respuesta_admin VARCHAR(500),
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @estadoRechazadaId INT;
    DECLARE @ordenId INT;
    DECLARE @usuarioId INT;
    
    BEGIN TRY
        BEGIN TRANSACTION;
        
        -- Validar que se proporcione el motivo
        IF @respuesta_admin IS NULL OR LTRIM(RTRIM(@respuesta_admin)) = ''
        BEGIN
            RAISERROR('El motivo del rechazo es requerido', 16, 1);
            ROLLBACK TRANSACTION;
            SET @accion = 0;
            RETURN;
        END
        
        -- Obtener ID del estado "RECHAZADA"
        SELECT @estadoRechazadaId = id 
        FROM estado_devolucion 
        WHERE codigo = 'RECHAZADA';
        
        IF @estadoRechazadaId IS NULL
        BEGIN
            RAISERROR('Estado RECHAZADA no encontrado', 16, 1);
            ROLLBACK TRANSACTION;
            SET @accion = 0;
            RETURN;
        END
        
        -- Obtener datos de la devolución
        SELECT 
            @ordenId = ordenId,
            @usuarioId = usuarioId
        FROM devolucion
        WHERE id = @devolucionId;
        
        -- Actualizar la devolución
        UPDATE devolucion
        SET 
            estadoDevolucionId = @estadoRechazadaId,
            fecha_respuesta = GETDATE(),
            respondido_por = @respondido_por,
            respuesta_admin = @respuesta_admin
        WHERE id = @devolucionId;
        
        -- Registrar en auditoría
        INSERT INTO auditoria (
            entidad,
            accion,
            descripcion,
            usuarioId,
            fecha
        )
        VALUES (
            'devolucion',
            'RECHAZAR',
            'Devolución #' + CAST(@devolucionId AS VARCHAR) + ' rechazada. Motivo: ' + @respuesta_admin,
            @respondido_por,
            GETDATE()
        );
        
        COMMIT TRANSACTION;
        SET @accion = 1;
        
        PRINT '✅ Devolución rechazada exitosamente';
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
            
        SET @accion = 0;
        PRINT '❌ Error al rechazar devolución: ' + ERROR_MESSAGE();
    END CATCH
END
GO
```

# [dbo].[sp_Rechazar_Producto]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Rechazar_Producto]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Rechazar_Producto]
    @productoId INT,
    @adminId INT,
    @motivo_rechazo VARCHAR(500) = NULL,
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        BEGIN TRANSACTION
        
        -- Verificar que el producto existe y está pendiente
        IF NOT EXISTS (SELECT 1 FROM producto WHERE id = @productoId AND estadoProductoId = 4)
        BEGIN
            SET @accion = 0;
            ROLLBACK TRANSACTION;
            PRINT '❌ El producto no existe o no está pendiente de aprobación';
            RETURN;
        END
        
        -- Actualizar producto a RECHAZADO
        UPDATE producto
        SET 
            estadoProductoId = 5,  -- RECHAZADO
            aprobado = 0,
            fecha_aprobacion = GETDATE(),
            aprobado_por = @adminId,
            motivo_rechazo = @motivo_rechazo
        WHERE id = @productoId;
        
        COMMIT TRANSACTION;
        SET @accion = 1;
        
        PRINT '✅ Producto ' + CAST(@productoId AS VARCHAR) + ' rechazado';
        
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
        
        SET @accion = 0;
        
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE();
        PRINT '❌ Error al rechazar producto: ' + @ErrorMessage;
        RAISERROR (@ErrorMessage, 16, 1);
    END CATCH
END
GO
```

# [dbo].[sp_Registrar_Auditoria]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registrar_Auditoria]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Registrar_Auditoria]
    @usuario_id INT = NULL,
    @usuario_email VARCHAR(255) = NULL,
    @usuario_nombre VARCHAR(255) = NULL,
    @usuario_rol VARCHAR(50) = NULL,
    @tipo_operacion VARCHAR(50),
    @entidad VARCHAR(100),
    @entidad_id INT = NULL,
    @accion VARCHAR(255),
    @tabla_afectada VARCHAR(100) = NULL,
    @valores_anteriores NVARCHAR(MAX) = NULL,
    @valores_nuevos NVARCHAR(MAX) = NULL,
    @ip_address VARCHAR(50) = NULL,
    @user_agent NVARCHAR(500) = NULL,
    @endpoint VARCHAR(255) = NULL,
    @metodo_http VARCHAR(10) = NULL,
    @exito BIT = 1,
    @codigo_error VARCHAR(50) = NULL,
    @mensaje_error NVARCHAR(MAX) = NULL,
    @datos_adicionales NVARCHAR(MAX) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Verificar si la auditoría está activa para esta entidad
    DECLARE @auditar BIT = 1;
    
    SELECT @auditar = CASE 
        WHEN @tipo_operacion = 'INSERT' THEN auditar_insert
        WHEN @tipo_operacion = 'UPDATE' THEN auditar_update
        WHEN @tipo_operacion = 'DELETE' THEN auditar_delete
        WHEN @tipo_operacion = 'SELECT' THEN auditar_select
        ELSE 1
    END
    FROM AuditConfiguration
    WHERE entidad = @entidad AND activo = 1;
    
    -- Si no existe configuración, auditar por defecto
    IF @auditar IS NULL SET @auditar = 1;
    
    IF @auditar = 1
    BEGIN
        INSERT INTO AuditLog (
            usuario_id, usuario_email, usuario_nombre, usuario_rol,
            tipo_operacion, entidad, entidad_id, accion, tabla_afectada,
            valores_anteriores, valores_nuevos,
            ip_address, user_agent, endpoint, metodo_http,
            exito, codigo_error, mensaje_error, datos_adicionales
        )
        VALUES (
            @usuario_id, @usuario_email, @usuario_nombre, @usuario_rol,
            @tipo_operacion, @entidad, @entidad_id, @accion, @tabla_afectada,
            @valores_anteriores, @valores_nuevos,
            @ip_address, @user_agent, @endpoint, @metodo_http,
            @exito, @codigo_error, @mensaje_error, @datos_adicionales
        );
    END
END
GO
```

# [dbo].[sp_Registrar_Devolucion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registrar_Devolucion]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- 5. SP: Registrar Nueva Devolución
-- ============================================
CREATE   PROCEDURE [dbo].[sp_Registrar_Devolucion]
    @ordenId INT,
    @usuarioId INT,
    @motivo VARCHAR(100),
    @tipo_devolucion VARCHAR(20),
    @descripcion_detallada VARCHAR(1000),
    @url_foto_1 VARCHAR(500) = NULL,
    @url_foto_2 VARCHAR(500) = NULL,
    @url_foto_3 VARCHAR(500) = NULL,
    @accion BIT OUTPUT,
    @devolucion_id INT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        -- Validar que la orden existe y está verificada
        IF NOT EXISTS (
            SELECT 1 FROM Orden o
            INNER JOIN estado_orden eo ON o.estadoOrdenId = eo.id
            WHERE o.id = @ordenId 
            AND eo.codigo = 'PAGO_VERIFICADO'
        )
        BEGIN
            SET @accion = 0;
            SET @devolucion_id = -1;
            RETURN;
        END
        
        -- Validar que no han pasado más de 15 días
        IF EXISTS (
            SELECT 1 FROM Orden
            WHERE id = @ordenId
            AND DATEDIFF(DAY, fecha, GETDATE()) > 15
        )
        BEGIN
            SET @accion = 0;
            SET @devolucion_id = -2; -- Código especial: fuera de plazo
            RETURN;
        END
        
        -- Validar que no existe una devolución activa para esta orden
        IF EXISTS (
            SELECT 1 FROM devolucion
            WHERE ordenId = @ordenId 
            AND activo = 1
            AND estadoDevolucionId NOT IN (
                SELECT id FROM estado_devolucion 
                WHERE codigo IN ('RECHAZADA', 'COMPLETADO', 'CANCELADO')
            )
        )
        BEGIN
            SET @accion = 0;
            SET @devolucion_id = -3; -- Código especial: ya existe devolución activa
            RETURN;
        END
        
        -- Insertar la devolución
        INSERT INTO devolucion (
            ordenId,
            usuarioId,
            estadoDevolucionId,
            motivo,
            tipo_devolucion,
            descripcion_detallada,
            url_foto_1,
            url_foto_2,
            url_foto_3,
            fecha_solicitud,
            activo
        )
        VALUES (
            @ordenId,
            @usuarioId,
            1, -- PENDIENTE
            @motivo,
            @tipo_devolucion,
            @descripcion_detallada,
            @url_foto_1,
            @url_foto_2,
            @url_foto_3,
            GETDATE(),
            1
        );
        
        SET @devolucion_id = SCOPE_IDENTITY();
        SET @accion = 1;
    END TRY
    BEGIN CATCH
        SET @accion = 0;
        SET @devolucion_id = 0;
        PRINT 'Error: ' + ERROR_MESSAGE();
    END CATCH
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registrar_Intento_Fallido]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- SP: Registrar Intento Fallido
-- ============================================
CREATE   PROCEDURE [dbo].[sp_Registrar_Intento_Fallido]
    @email VARCHAR(255),
    @max_intentos INT = 5,
    @tiempo_bloqueo_minutos INT = 15
AS
BEGIN
    DECLARE @intentos INT;
    DECLARE @bloqueado_hasta DATETIME;
    
    -- Incrementar contador de intentos
    UPDATE Usuario
    SET intentos_fallidos = ISNULL(intentos_fallidos, 0) + 1,
        fecha_ultimo_intento = GETDATE()
    WHERE email = @email;
    
    -- Obtener cantidad de intentos
    SELECT @intentos = intentos_fallidos
    FROM Usuario
    WHERE email = @email;
    
    -- Si excede el límite, bloquear cuenta
    IF @intentos >= @max_intentos
    BEGIN
        SET @bloqueado_hasta = DATEADD(MINUTE, @tiempo_bloqueo_minutos, GETDATE());
        
        UPDATE Usuario
        SET bloqueado_hasta = @bloqueado_hasta
        WHERE email = @email;
        
        PRINT 'Cuenta bloqueada hasta: ' + CONVERT(VARCHAR, @bloqueado_hasta, 120);
    END
    
    SELECT 
        @intentos as intentos_fallidos,
        @max_intentos as max_intentos,
        @bloqueado_hasta as bloqueado_hasta;
END;
GO
```

# [dbo].[sp_Registrar_Item]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registrar_Item]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_Registrar_Item]
@estadoItemId INT = 1,
@usuarioId INT,
@productoId INT,
@ordenId INT = NULL,
@cantidad INT,
@fecha DATETIME,
@accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        -- ✅ Verificar si el item YA existe en el carrito del usuario
        IF EXISTS (
            SELECT 1 
            FROM Items 
            WHERE usuarioId = @usuarioId 
            AND productoId = @productoId 
            AND estadoItemId = 1  -- En carrito
            AND ordenId IS NULL
        )
        BEGIN
            -- Si existe, incrementar cantidad
            UPDATE Items 
            SET cantidad = cantidad + @cantidad,
                fecha = @fecha
            WHERE usuarioId = @usuarioId 
            AND productoId = @productoId 
            AND estadoItemId = 1
            AND ordenId IS NULL;
            
            SET @accion = 2; -- ✅ Código 2 = Ya existía, se actualizó cantidad
        END
        ELSE
        BEGIN
            -- Si no existe, insertar nuevo
            INSERT INTO Items (estadoItemId, usuarioId, productoId, ordenId, cantidad, fecha)
            VALUES (@estadoItemId, @usuarioId, @productoId, @ordenId, @cantidad, @fecha);
            
            SET @accion = 1; -- ✅ Código 1 = Nuevo item agregado
        END
    END TRY
    BEGIN CATCH
        SET @accion = 0; -- ✅ Código 0 = Error
        PRINT ERROR_MESSAGE();
    END CATCH
END
GO
```

# [dbo].[sp_Registrar_Usuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Registrar_Usuario]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Crear el procedimiento CORREGIDO
CREATE PROCEDURE [dbo].[sp_Registrar_Usuario]
    @estadoUsuarioId INT = NULL,
    @rolId INT,
    @nombre VARCHAR(50),
    @apellido VARCHAR(50),
    @email VARCHAR(50),
    @password VARCHAR(255),
    @maxintentos INT,
    @intentosfallidos INT,
    @accion BIT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    BEGIN TRY
        BEGIN TRANSACTION
        
        DECLARE @nuevoUsuarioId INT;
        
        -- 🔹 CASO 1: Vendedor registrándose públicamente (rolId = 3)
        IF @rolId = 3 AND (@estadoUsuarioId IS NULL OR @estadoUsuarioId = 4)
        BEGIN
            -- Estado: PENDIENTE DE APROBACIÓN
            SET @estadoUsuarioId = 4;
            
            -- Insertar usuario
            INSERT INTO usuario(estadoUsuarioId, rolId, nombre, apellido, email, password, maxintentos, intentosfallidos, aprobado)
            VALUES(@estadoUsuarioId, @rolId, @nombre, @apellido, @email, @password, @maxintentos, @intentosfallidos, 0);
            
            SET @nuevoUsuarioId = SCOPE_IDENTITY();
            PRINT '✅ Usuario vendedor creado (PENDIENTE): ' + CAST(@nuevoUsuarioId AS VARCHAR);
            
            -- ❌ NO crear registro en vendedor aquí
            -- El registro se crea SOLO cuando se aprueba (sp_Aprobar_Usuario)
        END
        -- 🔹 CASO 2: Vendedor creado desde el panel admin con estado ACTIVO
        ELSE IF @rolId = 3 AND @estadoUsuarioId = 1
        BEGIN
            -- Estado: ACTIVO (creado por admin)
            
            -- Insertar usuario
            INSERT INTO usuario(estadoUsuarioId, rolId, nombre, apellido, email, password, maxintentos, intentosfallidos, aprobado, aprobado_por, fecha_aprobacion)
            VALUES(@estadoUsuarioId, @rolId, @nombre, @apellido, @email, @password, @maxintentos, @intentosfallidos, 1, NULL, GETDATE());
            
            SET @nuevoUsuarioId = SCOPE_IDENTITY();
            PRINT '✅ Usuario vendedor creado (ACTIVO desde admin): ' + CAST(@nuevoUsuarioId AS VARCHAR);
            
            -- ✅ CREAR REGISTRO EN VENDEDOR AUTOMÁTICAMENTE
            INSERT INTO vendedor (
                usuarioId,
                estadoUsuarioId,
                nombre_comercial,
                descripcion,
                telefono,
                fecha_registro
            )
            VALUES (
                @nuevoUsuarioId,
                1, -- Activo
                @nombre + ' ' + @apellido, -- Nombre comercial por defecto
                'Vendedor creado desde panel administrativo',
                '', -- Teléfono vacío por defecto
                GETDATE()
            );
            
            PRINT '✅ Registro de vendedor creado automáticamente para usuario: ' + CAST(@nuevoUsuarioId AS VARCHAR);
        END
        -- 🔹 CASO 3: Cualquier otro rol (Admin, Cliente, etc.)
        ELSE
        BEGIN
            SET @estadoUsuarioId = ISNULL(@estadoUsuarioId, 1); -- Activo por defecto
            
            -- Insertar usuario
            INSERT INTO usuario(estadoUsuarioId, rolId, nombre, apellido, email, password, maxintentos, intentosfallidos, aprobado)
            VALUES(@estadoUsuarioId, @rolId, @nombre, @apellido, @email, @password, @maxintentos, @intentosfallidos, 1);
            
            SET @nuevoUsuarioId = SCOPE_IDENTITY();
            PRINT '✅ Usuario creado (rol ' + CAST(@rolId AS VARCHAR) + '): ' + CAST(@nuevoUsuarioId AS VARCHAR);
        END
        
        COMMIT TRANSACTION;
        SET @accion = 1;
        
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;
        
        SET @accion = 0;
        
        DECLARE @ErrorMessage NVARCHAR(4000) = ERROR_MESSAGE();
        PRINT '❌ Error en sp_Registrar_Usuario: ' + @ErrorMessage;
        RAISERROR (@ErrorMessage, 16, 1);
    END CATCH
END
GO
```

# [dbo].[sp_reporte_vendedor]

```sql
/****** Object:  StoredProcedure [dbo].[sp_reporte_vendedor]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_reporte_vendedor]
@vendedorId INT
AS
BEGIN
    SET NOCOUNT ON;
    
    SELECT 
        v.id as vendedor_id,
        v.nombre_comercial,
        COUNT(DISTINCT p.id) as total_productos_activos,
        ISNULL(SUM(i.cantidad * p.precio_ahora), 0) as total_ventas,
        COUNT(DISTINCT o.id) as total_ordenes,
        ISNULL(AVG(o.total), 0) as promedio_por_orden,
        MIN(o.fecha) as fecha_primera_venta,
        MAX(o.fecha) as fecha_ultima_venta
    FROM vendedor v
    LEFT JOIN producto p ON p.vendedorId = v.id AND p.estadoProductoId IN (1, 2)
    LEFT JOIN Items i ON i.productoId = p.id AND i.estadoItemId = 2
    LEFT JOIN Orden o ON o.id = i.ordenId AND o.estadoOrdenId IN (4, 5)
    WHERE v.id = @vendedorId
    GROUP BY v.id, v.nombre_comercial
END
GO
```

# [dbo].[sp_ventas_por_mes]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ventas_por_mes]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- SP 2: VENTAS POR MES (ÚLTIMOS 12 MESES)
-- ============================================
CREATE PROCEDURE [dbo].[sp_ventas_por_mes]
AS
BEGIN
    SELECT 
        MONTH(o.fecha) as mes,
        YEAR(o.fecha) as año,
        DATENAME(MONTH, o.fecha) as mes_nombre,
        SUM(o.total) as total,
        COUNT(o.id) as cantidad_ordenes
    FROM orden o
    WHERE o.fecha >= DATEADD(MONTH, -12, GETDATE())
    GROUP BY YEAR(o.fecha), MONTH(o.fecha), DATENAME(MONTH, o.fecha)
    ORDER BY año DESC, mes DESC
END
GO
```

# [dbo].[sp_ventas_semanales]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ventas_semanales]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_ventas_semanales]
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Crear tabla temporal con los últimos 7 días
    DECLARE @FechaInicio DATE = DATEADD(DAY, -6, CAST(GETDATE() AS DATE));
    DECLARE @FechaFin DATE = CAST(GETDATE() AS DATE);
    
    ;WITH Dias AS (
        SELECT @FechaInicio AS fecha
        UNION ALL
        SELECT DATEADD(DAY, 1, fecha)
        FROM Dias
        WHERE fecha < @FechaFin
    )
    SELECT 
        d.fecha,
        DATENAME(WEEKDAY, d.fecha) as dia_nombre,
        FORMAT(d.fecha, 'dd/MM') as fecha_corta,
        ISNULL(SUM(o.total), 0) as total,
        COUNT(o.id) as cantidad_ordenes
    FROM Dias d
    LEFT JOIN Orden o ON CAST(o.fecha AS DATE) = d.fecha 
        AND o.estadoOrdenId IN (3, 4, 5) -- Solo órdenes con comprobante o verificadas
    GROUP BY d.fecha
    ORDER BY d.fecha ASC
    OPTION (MAXRECURSION 7);
END
GO
```

# [dbo].[sp_Verificar_PuedeDevolver]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Verificar_PuedeDevolver]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_Verificar_PuedeDevolver]
    @ordenId INT,
    @puede_devolver BIT OUTPUT,
    @motivo_no_puede VARCHAR(200) OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @dias_transcurridos INT;
    DECLARE @estado_codigo VARCHAR(30);
    DECLARE @estado_devolucion_activa VARCHAR(30);

    IF NOT EXISTS (SELECT 1 FROM Orden WHERE id = @ordenId)
    BEGIN
        SET @puede_devolver = 0;
        SET @motivo_no_puede = 'La orden no existe';
        RETURN;
    END

    SELECT @estado_codigo = eo.codigo
    FROM Orden o
    INNER JOIN estado_orden eo ON o.estadoOrdenId = eo.id
    WHERE o.id = @ordenId;

    IF @estado_codigo != 'PAGO_VERIFICADO'
    BEGIN
        SET @puede_devolver = 0;
        SET @motivo_no_puede = 'Solo se pueden devolver órdenes con pago verificado';
        RETURN;
    END

    SELECT @dias_transcurridos = DATEDIFF(DAY, fecha, GETDATE())
    FROM Orden WHERE id = @ordenId;

    IF @dias_transcurridos > 15
    BEGIN
        SET @puede_devolver = 0;
        SET @motivo_no_puede = 'El plazo de devolución (15 días) ha expirado';
        RETURN;
    END

    -- Obtener el estado de la devolución activa si existe
    SELECT TOP 1 @estado_devolucion_activa = ed.codigo
    FROM devolucion d
    INNER JOIN estado_devolucion ed ON d.estadoDevolucionId = ed.id
    WHERE d.ordenId = @ordenId
    AND d.activo = 1
    AND ed.codigo NOT IN ('RECHAZADA', 'COMPLETADO', 'CANCELADO');

    IF @estado_devolucion_activa IS NOT NULL
    BEGIN
        SET @puede_devolver = 0;
        SET @motivo_no_puede = CASE @estado_devolucion_activa
            WHEN 'PENDIENTE'   THEN 'Tu solicitud de devolución está pendiente de revisión'
            WHEN 'APROBADO'    THEN 'Tu devolución ya fue aprobada'
            WHEN 'EN_PROCESO'  THEN 'Tu devolución está en proceso'
            ELSE 'Ya existe una solicitud de devolución activa para esta orden'
        END;
        RETURN;
    END

    SET @puede_devolver = 1;
    SET @motivo_no_puede = NULL;
END
GO
```

# [dbo].[sp_Verificar_Token_Recuperacion]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Verificar_Token_Recuperacion]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- SP: Verificar Token de Recuperación
-- ============================================
CREATE PROCEDURE [dbo].[sp_Verificar_Token_Recuperacion]
    @token VARCHAR(100)
AS
BEGIN
    SELECT 
        id,
        email,
        token,
        fecha_creacion,
        fecha_expiracion,
        usado,
        fecha_uso
    FROM Tokens_Recuperacion
    WHERE token = @token
      AND usado = 0
      AND fecha_expiracion > GETDATE();
END;
GO
```

# [dbo].[sp_Verificar_Usuario_Bloqueado]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Verificar_Usuario_Bloqueado]    Script Date: 10/03/2026 7:13:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- ============================================
-- SP: Verificar si Usuario está Bloqueado
-- ============================================
CREATE   PROCEDURE [dbo].[sp_Verificar_Usuario_Bloqueado]
    @email VARCHAR(255)
AS
BEGIN
    SELECT 
        CASE 
            WHEN bloqueado_hasta IS NOT NULL AND bloqueado_hasta > GETDATE() 
            THEN 1 
            ELSE 0 
        END as esta_bloqueado,
        bloqueado_hasta,
        intentos_fallidos,
        DATEDIFF(MINUTE, GETDATE(), ISNULL(bloqueado_hasta, GETDATE())) as minutos_restantes
    FROM Usuario
    WHERE email = @email;
END;
GO
```
